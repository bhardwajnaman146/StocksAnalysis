```python
df['%age Lumpsum'] = df['%age Lumpsum'].str.rstrip('%').astype('float')
df['%age 1% Dip'] = df['%age 1% Dip'].str.rstrip('%').astype('float')
df['%age 2% dip'] = df['%age 2% dip'].str.rstrip('%').astype('float')
df['%age for -2% and -1% dip'] = df['%age for -2% and -1% dip'].str.rstrip('%').astype('float')
df['%age for sip'] = df['%age for sip'].str.rstrip('%').astype('float')
df.head()
```

```python
# Changing the value of spot %age column 1st row just to normalize everything 
df['spot %age'][0] = 1
df.head()
```

```python
# plotting the chart
px.line(df, x = 'date', y=["%age Lumpsum","%age 1% Dip","%age 2% dip","%age for -2% and -1% dip","%age for sip"], title='BOOM')
```
